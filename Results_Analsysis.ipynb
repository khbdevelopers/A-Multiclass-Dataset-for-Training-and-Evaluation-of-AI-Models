```python
#Accuracy
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Dummy results (replace with actual results)
models = ['LR', 'RF', 'XGB', 'SVM', 'KNN']
accuracy = [99.10, 99.33, 99.37, 99, 96.29]
precision = [99.10, 99.33, 99.37, 99, 96.29]
recall = [99.10, 99.33, 99.37, 99, 96.29]
f1_score = [99.10, 99.33, 99.37, 99, 96.29]
bar_width = 0.2
index = np.arange(len(models))

fig, ax = plt.subplots(figsize=(16, 6))

bar1 = ax.bar(index - bar_width * 1.5, accuracy, bar_width, label='Accuracy', color='#8C564B')
bar2 = ax.bar(index - bar_width * 0.5, precision, bar_width, label='Precision', color='lightgreen')
bar3 = ax.bar(index + bar_width * 0.5, recall, bar_width, label='Recall', color='#7F7F7F')
bar4 = ax.bar(index + bar_width * 1.5, f1_score, bar_width, label='F1-Score', color='lightblue')

# Add accuracy values on top of the bars at the center
for bars, accuracies in zip([bar1, bar2, bar3, bar4], [accuracy, precision, recall, f1_score]):
    for i, acc in enumerate(accuracies):
        if acc == 100:  # Check if accuracy is 100
            ax.text(bars[i].get_x() + bars[i].get_width() / 2, acc + 0.01, f'{int(acc)}',
                    ha='center', va='bottom', color='black')
        else:
            ax.text(bars[i].get_x() + bars[i].get_width() / 2, acc + 0.01, f'{acc:.2f}',
                    ha='center', va='bottom', color='black')

# Place the legend at the bottom left
ax.legend(loc='lower left')

plt.xlabel('Machine Learning Models')
plt.ylabel('Scores')
plt.title('Models Performance in Binary Classification')
plt.xticks(index, models)
plt.savefig("/content/drive/MyDrive/ML/Thesis/Dataset/Graphs/Models/binary-comare.pdf")   # <-- save as PDF
plt.show()
```
